# Movie Recommendation System 

### There are 3 types of recommender system 
- Content based : based on the content.
- Collaborative filtering: based on the similarity of customers ,similarity of interest.
- Hybrid : mix of above two.

```python
import numpy as np 
import pandas as pd 
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.dropna(inplace=True)
```

```python
#check duplicate movies
movies.duplicated().sum()
```

```python
import ast 
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
    
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']= movies['keywords'].apply(convert)
movies.head(1)
```

```python
def convert1(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<3:
            l.append(i['character'])
        counter+=1
    return l        
```

```python
def convert2(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(convert2) 
movies.head(1)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head(1)
```

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
columns = ['genres','cast','crew','keywords','overview']
movies_new=movies.drop(columns=columns)
movies_new
```

```python
movies_new['tags']=movies_new['tags'].apply(lambda x:" ".join(x))
```

```python
movies_new['tags'][0]
```

```python
new_df = movies_new[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

## STEMMING
### stemming is a technique done to fetch root words in different forms of word such as loved,loves,loving and just fetch the root word love.

```python
import nltk 
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags'].apply(stem)
```

## Text vectorisation 
### counter vectoriser is used to convert the string into vectors so that machine can learn 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector =cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# similarity is inverse of distance and we can get similarity by sklearn.metrics.pairwise import cosine_distance
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
